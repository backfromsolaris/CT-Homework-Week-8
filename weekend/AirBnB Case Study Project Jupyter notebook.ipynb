```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('./AB_NYC_2019.csv')
air_bnb.head(3)
```

```python
# 1. Which hosts are the busiest and why?

air_bnb.groupby(['host_id','host_name']).sum()[['number_of_reviews']].sort_values('number_of_reviews', ascending=False).head(5)


# Maya, Brooklyn & Breakfest -Len-, Danielle, Yasu/Akiko, and Brady of their respective neighborhoods shown below
# are the busiest hosts measured by total number of reviews
```

```python
# 2. How many neighborhood groups are available and which shows up the most?

air_bnb['appearances'] = air_bnb['id'].count()
air_bnb.groupby('neighbourhood_group').count().sort_values('appearances', ascending=False)[['appearances']]

# Manhattan has the most appearances of all neighbourhood groups within this dataset at 21,661
```

```python
# 3. Are private rooms the most popular in manhattan?
air_bnb = pd.read_csv('./AB_NYC_2019.csv')

air_bnb['room_type_count'] = air_bnb['room_type'].count()
air_bnb[air_bnb['neighbourhood_group'] == 'Manhattan'].set_index('room_type').groupby(['room_type', 'neighbourhood_group']).count()[['room_type_count']]

# Entire home/apt listings are the most popular in manhattan, not private rooms
```

```python
# 4. Which 5 hosts have the most bookings, based purely on reviews per month?
air_bnb = pd.read_csv('./AB_NYC_2019.csv')

air_bnb.groupby(['host_id', 'host_name']).count()[['reviews_per_month']].sort_values('reviews_per_month', ascending=False).head(6)

# top 5 hosts based on reviews per month shown below (6th shown to demonstrate lack of a tie)
```

```python
# 5. Which neighorhood group has the highest average price?

air_bnb.head(5)

air_bnb.groupby(['neighbourhood_group']).mean()[['price']].round(decimals=2).sort_values('price', ascending=False)

# Manhattan has the highest avg price
```

```python
# 6. Which neighborhood group has the highest total price?

air_bnb.groupby(['neighbourhood_group']).sum()[['price']].round(decimals=2).sort_values('price', ascending=False)

# Manhattan has the highest total price, when taking the sum of all listings within each neighbourhood group
```

```python
# 7. Which top 5 hosts have the highest total price?

air_bnb.groupby(['host_id', 'host_name']).sum()[['price']].sort_values('price', ascending=False).head(5)

# top 5 hosts determined by total price shown below
```

```python
# 8. Who currently has no (zero) availability with a review count of 100 or more?

hundo_reviews = air_bnb[air_bnb['number_of_reviews'] >= 100]
hundo_reviews[hundo_reviews['availability_365'] == 0].groupby(['host_id', 'host_name']).sum()[['number_of_reviews', 'availability_365']].sort_values('number_of_reviews')

# see results below: 147 hosts have zero avail with a review count at or above 100
```

```python
# 9. What host has the highest total of prices and where are they located?

air_bnb.groupby(['host_id', 'host_name', 'neighbourhood', 'neighbourhood_group']).sum()[['price']].sort_values('price', ascending=False)

# Sonder, located in Financial District, Manhattan, has the highest total of prices
```

```python
# 10. When did Danielle from Queens last receive a review?

danielle = air_bnb[air_bnb['host_name'] == 'Danielle']
danielle_queens = danielle[danielle['neighbourhood_group'] == 'Queens']
danielle_queens[['host_name', 'neighbourhood_group','last_review']].sort_values('last_review', ascending=False)

# Danielle from Queens received her last review on 7/8/2019
```
